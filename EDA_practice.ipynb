```python
visa_df=pd.read_csv('Visadataset.csv')
unique_col=visa_df['education_of_employee'].unique()
val_con=[]
for i in unique_col:
    con=visa_df['education_of_employee']==i
    val=len(visa_df[con])
    val_con.append(val)
    print(i,val)
```

```python
A=pd.DataFrame(zip(unique_col,val_con),columns=['education_of_employees','count'])
A
```

```python
A.to_csv('education_of_employees_df.csv',index=False)
```

```python
plt.figure(figsize=(5,5))
plt.bar('education_of_employees','count',data=A)
plt.show()
```

```python
plt.pie(val_con,explode=[0.1,0.1,0.1,0.1],labels=unique_col,autopct='%0.2f%%',startangle=180,radius=2)
plt.show()
```
