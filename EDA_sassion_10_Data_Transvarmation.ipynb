```python
from scipy.stats import boxcox
from scipy.stats import probplot

data = np.array([4, 2, 1, 3, 5, 6, 8, 10, 12])

plt.figure(figsize=(12, 6))
plt.subplot(1, 2, 1)
plt.hist(data, bins=10, color='blue', edgecolor='black')
plt.title('Original Data Distribution')

transformed_data, lambda_optimal = boxcox(data)

plt.subplot(1, 2, 2)
plt.hist(transformed_data, bins=10, color='green', edgecolor='black')
plt.title(f'Box-Cox Transformed Data (λ={lambda_optimal:.2f})')

plt.show()

print(f"Optimal λ: {lambda_optimal:.2f}")

plt.figure(figsize=(12, 6))
plt.subplot(1, 2, 1)
probplot(data, dist="norm", plot=plt)
plt.title('Q-Q Plot of Original Data')

plt.subplot(1, 2, 2)
probplot(transformed_data, dist="norm", plot=plt)
plt.title(f'Q-Q Plot of Box-Cox Transformed Data (λ={lambda_optimal:.2f})')

plt.show()
```
