```python
unique_cun=visa_df['continent'].unique()
unique_cun
```

```python
A=[]
for i in unique_cun:
    con=visa_df['continent']==i
    val=len(visa_df[con])
    A.append(val)
    print(f"{i}:{val}")
A
```

```python
pd.DataFrame(zip(unique_cun,A),columns=['continent','count'])
```

```python
countinent_df=pd.DataFrame(zip(unique_cun,A),columns=['continent','count'])
```

```python
countinent_df.to_csv('countinent_df',index=False)
```

**Bar chart**

- Bar chart is represantation of count w.r.t classes
- if we want plot bar chart we required two  columns
 - another column is Numarecal data column

```python
plt.figure(figsize=(8,5))
plt.bar('continent','count',data=countinent_df)
plt.title("bar chart")
plt.xlabel("continent")
plt.ylabel("count")
plt.savefig("barchart.jpg")
plt.show()
```
